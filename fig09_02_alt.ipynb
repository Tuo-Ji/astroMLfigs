<h2>Let's take the example from the dialog box on page 372.</h2>

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

X = np.random.random((100,2))
vals = (X[:,0] + X[:, 1] > 1)
y = vals.astype(int)

gnb = GaussianNB()
gnb.fit(X, y)
y_pred = gnb.predict(X)
```

```python
#print np.c_[y, y_pred]
tvals = y == y_pred
print tvals.astype(int)
complemen = (tvals == False)
n_wrong = len(tvals)-sum(tvals)
#print "there were %f3 "
print('There were {} misclassifications!'.format(n_wrong))
```

```python
#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)
ax.scatter(X[complemen, 0], X[complemen, 1], c='r', 
           cmap=plt.cm.binary, zorder=2, s=160, alpha=0.3)
ax.plot([0,1], [1, 0], '--r', lw=4)
#ax.contour(xx, yy, Z, [0.5], colors='k')

#ax.set_xlim(xlim)
#ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title("Misclassifications in Gaussian Naive Bayes")

plt.show()
```

<h1 align=center> Linear Discriminant Analysis and Quadratic Discriminant Analysis</h1>
<h2 align=center> from the dialog box on page 375</h2>

```python
from sklearn.lda import LDA
from sklearn.qda import QDA
```

we'll use the same $X$ and $y$ as above...

```python
#Okay, Python we're going to do some LDA here, get ready
lda = LDA()
qda = QDA()
#Okay, not that you've got that LDA model in your head, 
#train it on this data-- X is the list of vectors,
#y are the (binary) class labels
lda.fit(X, y)
qda.fit(X, y)
#predict the class labels for the same X vectors I just gave you.
y_pred_lda = lda.predict(X)
y_pred_qda = qda.predict(X)
```

find the miss classifications, as we did last time, except way more professional with the boolean logic!

```python
#print np.c_[y, y_pred]
lda_bi = (y != y_pred_lda) 
qda_bi = (y != y_pred_qda)
print(' LDA missclassified {} sources \n QDA missclassified {} sources'.format(sum(lda_bi), sum(qda_bi)))
```

```python
#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)
ax.scatter(X[lda_bi, 0], X[lda_bi, 1], c='r', 
           cmap=plt.cm.binary, zorder=2, s=160, alpha=0.3)
ax.scatter(X[qda_bi, 0], X[qda_bi, 1], c='b', 
           cmap=plt.cm.binary, zorder=2, s=160, alpha=0.3)
ax.plot([0,1], [1, 0], '--r', lw=4)
#ax.contour(xx, yy, Z, [0.5], colors='k')

#ax.set_xlim(xlim)
#ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title("Misclassifications in LDA and QDA")

plt.show()
```

<h3>What happens if *just ONE* source is *FLIPPED* from class 1 to 0, or 0 to 1?</h3>

```python
print y[0]
print X[0,]
y_orig = y[0]
```

```python
if y_orig == 0:
    y[0] = 1
else:
    y[0] = 0
```

```python
lda = LDA()
qda = QDA()
lda.fit(X, y)
qda.fit(X, y)
y_pred_lda = lda.predict(X)
y_pred_qda = qda.predict(X)
#print np.c_[y, y_pred]
lda_bi = (y != y_pred_lda) 
qda_bi = (y != y_pred_qda)
print(' LDA missclassified {} sources \n QDA missclassified {} sources'.format(sum(lda_bi), sum(qda_bi)))
#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)
ax.scatter(X[lda_bi, 0], X[lda_bi, 1], c='r', 
           cmap=plt.cm.binary, zorder=2, s=160, alpha=0.3)
ax.scatter(X[qda_bi, 0], X[qda_bi, 1], c='b', 
           cmap=plt.cm.binary, zorder=2, s=160, alpha=0.3)
ax.plot([0,1], [1, 0], '--r', lw=4)
#ax.contour(xx, yy, Z, [0.5], colors='k')

#ax.set_xlim(xlim)
#ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title("Misclassifications in LDA and QDA")

plt.show()
```

The result is not terrible: *life goes on* despite the obvious outlier.  This probably makes sense when you dig into how the classification actually works.  But barring that detailed explanation, the key idea is that the machine learning is extremizing some function, and the fact that 1% of the data points changed doesn't materially affect the optimization.  One cute thing to notice however... the errant data point pulled the classification of open circles *towards* the errant data point.  It makes sense.

<h2>Let's do all the generative classifications</h2>

```python
if y_orig == 0:
    y[0] = 1
else:
    y[0] = 0
```

```python
from astroML.classification import GMMBayes
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

gmmb = GMMBayes(3)
knc = KNeighborsClassifier(5)
logr = LogisticRegression(penalty='l2')
svm = LinearSVC(loss='l2')
dtree = DecisionTreeClassifier(max_depth=6)
rfor = RandomForestClassifier(10)
gboo = GradientBoostingClassifier()


gmmb.fit(X, y)
knc.fit(X, y)
logr.fit(X, y)
svm.fit(X, y)
dtree.fit(X, y)
rfor.fit(X, y)
gboo.fit(X, y)

y_pred_gmmb = gmmb.predict(X)
y_pred_knc = knc.predict(X)
y_pred_logr = logr.predict(X)
y_pred_svm = svm.predict(X)
y_pred_dtree = dtree.predict(X)
y_pred_rfor = rfor.predict(X)
y_pred_gboo = gboo.predict(X)
```

```python
#print np.c_[y, y_pred]

bgm = y_pred_gmmb != y
bk = y_pred_knc != y
bl = y_pred_logr != y
bs = y_pred_svm != y
bd = y_pred_dtree != y
br = y_pred_rfor != y
bgb = y_pred_gboo != y

print sum(bgm), sum(bk), sum(bl), sum(bs), sum(bd), sum(br), sum(bgb)
```

```python
fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)

#Real classifications
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, s=10)

#Classifications
ax.scatter(X[bgm, 0], X[bgm, 1], c='r', s=60, alpha=0.15)
ax.scatter(X[bk, 0], X[bk, 1], c='g', s=120, alpha=0.15)
ax.scatter(X[bl, 0], X[bl, 1], c='b', s=180, alpha=0.15)
ax.scatter(X[bs, 0], X[bs, 1], c='y', s=240, alpha=0.15)
ax.scatter(X[bd, 0], X[bd, 1], c='r', s=300, alpha=0.15)
ax.scatter(X[br, 0], X[br, 1], c='m', s=360, alpha=0.15)
ax.scatter(X[bgb, 0], X[bgb, 1], c='c', s=420, alpha=0.15)

ax.plot([0,1], [1, 0], '--r', lw=4)
#ax.contour(xx, yy, Z, [0.5], colors='k')

#ax.set_xlim(xlim)
#ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title("Misclassifications, see chapter 9")

plt.show()
```
